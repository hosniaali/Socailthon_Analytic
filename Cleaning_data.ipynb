```python
dash_data4=pd.read_excel(r'C:\Users\najla\Old_OneDrive_Stuff\Desktop\SocialThon\Data\dash_data2.xlsx')
dash_data4
```

```python
dash_data4['less_7']=dash_data4['أقل من 7 سنوات']
dash_data4['more_than_60']=dash_data4['أكثر من 60 سنة']
dash_data4['from_19_35']=dash_data4['من 19 - 35 سنة']
dash_data4['from_36_60']=dash_data4['من 36 - 60 سنة']
dash_data4['from_7_18']=dash_data4['من 7 - 18 سنة']
dash_data4['total_num_of_institution']=dash_data4['less_7']+dash_data4['more_than_60']+dash_data4['from_19_35']+dash_data4['from_36_60']+dash_data4['from_7_18']
```

```python
dash_data4['less_7']=round((dash_data4['less_7']/dash_data4['total_num_of_institution'])*100,2)
dash_data4['more_than_60']=round((dash_data4['more_than_60']/dash_data4['total_num_of_institution'])*100,2)
dash_data4['from_19_35']=round((dash_data4['from_19_35']/dash_data4['total_num_of_institution'])*100,2)
dash_data4['from_36_60']=round((dash_data4['from_36_60']/dash_data4['total_num_of_institution'])*100,2)
dash_data4['from_7_18']=round((dash_data4['from_7_18']/dash_data4['total_num_of_institution'])*100,2)
```

```python
dash_data4
```

```python
dash_data4.columns
```

```python
dash_data4['beneficiaries_to_institution']=round(dash_data4['num_of_beneficiaries']/dash_data4['total_num_of_institution'])
```
